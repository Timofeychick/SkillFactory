```python
def factorial(num = 0):
    if not num:
        return 1
    return num * factorial(num - 1)
```

```python
from graphviz import Digraph

def draw_factorial_tree(n):
    dot = Digraph()
    
    def add_node(n):
        dot.node(str(n), f'factorial({n})')
        if n > 0:
            dot.edge(str(n), str(n-1))
            add_node(n - 1)
        else:
            dot.node("base", "return 1", shape="box")
            dot.edge(str(n), "base")

    add_node(n)
    return dot

draw_factorial_tree(4)
```

```python
def combination (n, k):
    nk = n - k

    def factorial(x):
        if not x:
            return 1
        return x * factorial(x - 1)
    
    return factorial(n) / (factorial(nk) * factorial(k))
```

```python
def fib(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    return fib(n - 1) + fib(n - 2)
```

```python
def print_dict(input_data, level=0):
    # Если input_data — словарь 
    if type(input_data) is dict:
        # Создаём цикл по ключам словаря
        for key in input_data:
            # Выводим ключ в формате "<пробелы> <имя ключа> ->"
            print('  ' * level + f'{key} ->')
            #print('  ' * level + '{} ->'.format(key))
            # Повторяем те же операции для каждого значения словаря
            print_dict(input_data[key], level=level+1)  
    else: # В противном случае
        # Выводим значения в формате "<пробелы> <значения>"
        print('  ' * level + str(input_data))
```

```python
def equalize_lengths(lst):
    max_length = len(max(lst, key = len))
    corr_lst = lst.copy()
    corr_lst.sort(key = len)
    corr_lst = corr_lst[::-1]
    
    for i in range (len(corr_lst)):
        if len(corr_lst[i]) < max_length:
            corr_lst[i] += ('_' * (max_length - len(corr_lst[i])))
    return corr_lst

print(equalize_lengths(['qweasdqweas', 'q', 'rteww', 'ewqqqqq']))
```

```python
def more_than_n(lst, n):
    more_then = list()
    if len(lst) == 0:
        return more_then
    else:
        for num in lst:
            if not (-1 * n < num < n):
                more_then.append(num)
        return more_then
```

```python
def even_numbers_in_matrix(matrix):
    counter = 0
    for lst in matrix:
        for elem in lst:
            if elem % 2 == 0:
                counter += 1

    return counter
```

```python
def matrix_sum(matrix1, matrix2):

    return_matrix = matrix1.copy()

    if len(matrix1) == len(matrix2) and len(matrix1[1]) == len(matrix2[1]):
        
        for i, lst in enumerate(matrix1):
            for j, elem in enumerate(lst):
                return_matrix[i][j] = elem + matrix2[i][j]
        return return_matrix
    else:
        print("Error! Matrices dimensions are different!")
        return None
```

# 1 
## - итог: словарь со структурой {называние сыра: жирность}

```python
def filter_by_fat(data, number):
    result = {}

    for key, value in data.items():
        if value[2] <= number:
            result[key] = value[2]
    
    return result

print(filter_by_fat(cheese_data, 25))
```

# 2
## - посчитать сколько деняк заработает сыроварня изходя из цены за 100 гр и максимального количества кг в месяц

```python
def count_money(data):
    result = 0

    for key, value in data.items():
        price = data[key][0]
        volume = data[key][1]
        result += (price * volume * 10)
    
    return result

print(count_money(cheese_data))
```

```python
def find_cheese_type(data, cheese_type=''):
    result = []

    for key, value in data.items():
        if value[-1] == cheese_type:
            result.append(key)
    
    return result

print(find_cheese_type(cheese_data, cheese_type='мягкий'))
```
